```python
import numpy
import matplotlib
```

```python
file_path = './cosySpectrum.pt2'
with open(file_path, 'rb') as f:
    binary_data = f.read()

# Truncate the first 3386 bytes (the header)
header_size = 3386
binary_data_truncated = binary_data[header_size:]
binary_data=binary_data_truncated
```

```python
import numpy as np

expected_size_bytes = 8192 * 128 * np.dtype(np.float32).itemsize

if len(binary_data) != expected_size_bytes:
    print(f"Error: Binary data size mismatch. Expected {expected_size_bytes} bytes, but got {len(binary_data)} bytes.")
else:
    data_array = np.frombuffer(binary_data, dtype=np.float32)
    data_2d = data_array.reshape((128,8192))
    print("Data reshaped successfully to 8192x128.")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# The data_2d from cosySpectrum.pt2 is already real, no need to extract real part

# Set intensity threshold
threshold = 500
masked_data = np.ma.masked_where(data_2d < threshold, data_2d)

# Plot the data using a contour plot with the threshold applied
plt.figure(figsize=(10, 8))
plt.contourf(masked_data[20:80,3000:5000], levels=50, cmap='viridis')
plt.colorbar(label='Intensity')
plt.title('Contour Plot of COSY Spectrum (Thresholded)')
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.show()
```

## Read binary data

### Subtask:
Read the raw binary data from `/content/COSY_expert/cosyFID.pt2`.


```python
file_path_fid = './cosyFID.pt2'
with open(file_path_fid, 'rb') as f:
    binary_data_fid = f.read()
```

## Summary:

### Data Analysis Key Findings

* A 2D complex Fourier transform was successfully applied to the initial 2D NMR data.
* The imaginary part of the transformed data was rejected, and a new complex dataset with dimensions of 4096x128 was created by using the even columns of the real part as the real components and the odd columns as the imaginary components.
* A Fourier transform was performed on the second dimension (with 128 complex points) of the newly constructed complex data.

### Insights or Next Steps

* The sequential application of Fourier transforms and data restructuring has prepared the NMR data for further analysis, such as peak picking or visualization.

```python
header_size_fid = 3386
binary_data_truncated_fid = binary_data_fid[header_size_fid:]
```

```python
# Define dimensions as variables
dim2_size = 8192
dim1_size = 128
```

```python
import numpy as np

data_array_fid = np.frombuffer(binary_data_truncated_fid, dtype=np.complex64)
data_2d_fid = data_array_fid.reshape((dim1_size, dim2_size))
```

```python
print(f"Size of data_array_fid: {data_array_fid.size}")
```

```python
import numpy as np

expected_size_bytes_fid = dim1_size * dim2_size * np.dtype(np.complex64).itemsize

if len(binary_data_truncated_fid) != expected_size_bytes_fid:
    print(f"Error: Binary data size mismatch for FID. Expected {expected_size_bytes_fid} bytes, but got {len(binary_data_truncated_fid)} bytes.")
else:
    data_array_fid = np.frombuffer(binary_data_truncated_fid, dtype=np.complex64)
    data_2d_fid = data_array_fid.reshape((dim1_size, dim2_size))
    print("FID data reshaped successfully.")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Extract the real part of the data_2d_fid and transpose it
real_data_2d_fid = np.real(data_2d_fid)

plt.figure(figsize=(10, 8))

# Create a stack plot
offset = np.max(real_data_2d_fid) * 0.5 # Adjust offset as needed
for i in range(10):
    plt.plot(real_data_2d_fid[i, :] + i * offset, color='black', linewidth=0.5)

plt.title('Stack Plot of Real Part of FID Data (Transposed)')
plt.xlabel('Dimension 1') # Swapped labels after transpose
plt.ylabel('Dimension 2 (offset)') # Swapped labels after transpose
plt.yticks([]) # Hide y-axis ticks as the stacking provides the dimension information
plt.show()
```

# Task
Process the 2D NMR data from "/content/COSY_expert/cosyFID.pt2" by applying a 2D complex Fourier transform, restructuring the data to create a new complex signal, and then applying a Fourier transform to the second dimension.

## Apply 2d complex fourier transform

### Subtask:
Perform a 2D complex Fourier transform on `data_2d_fid`.


**Reasoning**:
Apply a 2D Fourier transform to the `data_2d_fid` array using the `fftpack.fft2` function and store the result in a new variable called `transformed_data`.



```python
from scipy import fftpack

# Apply 1D Fourier transform to the first dimension (axis 0, size 4096)
transformed_data = fftpack.fft(data_2d_fid, axis=0)

# Apply fftshift to the first dimension
transformed_data = fftpack.fftshift(transformed_data, axes=0)
```

```python
from scipy import fftpack

# Apply Fourier transform to the second dimension (axis 1, size 128) of the new complex data
final_transformed_data = fftpack.fft(transformed_data, axis=1)

# Apply fftshift to the second dimension
#final_transformed_data = fftpack.fftshift(final_transformed_data, axes=1)

print("Fourier transform applied to the second dimension with fftshift.")
```

## Restructure data

### Subtask:
Reject the imaginary part of the transformed data. Then, create a new complex signal by taking every second row (starting from the first) in the second dimension (originally 256 points) as the real part and every second row (starting from the second) as the imaginary part. This will result in a new 2D array with dimensions 4096x128 complex points.


**Reasoning**:
Extract the real part of the transformed data and create a new complex array by combining even and odd indexed columns as real and imaginary parts, respectively.



```python
real_transformed_data = np.real(transformed_data)
new_complex_data = real_transformed_data[:, ::2] + 1j * real_transformed_data[:, 1::2]
```

## Apply fourier transform to second dimension

### Subtask:
Perform a Fourier transform on the second dimension (the dimension with 128 complex points) of the newly constructed complex data.


**Reasoning**:
Apply the Fourier transform to the second dimension of the `new_complex_data` array.



```python
from scipy import fftpack

final_transformed_data = fftpack.fft(new_complex_data, axis=1)
final_transformed_data = fftpack.fftshift(final_transformed_data, axes=1)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Extract the real part of the final transformed data
real_final_transformed_data = np.real(final_transformed_data)

# Set intensity threshold
threshold = 500
masked_data = np.ma.masked_where(real_final_transformed_data < threshold, real_final_transformed_data)

# Plot the real part using a contour plot with the threshold applied
plt.figure(figsize=(10, 8))
plt.contourf(masked_data[20:80,1500:2500], levels=50, cmap='viridis')
plt.colorbar(label='Intensity')
plt.title('Contour Plot of Real Part of Processed FID Data (Thresholded)')
plt.xlabel('Dimension 2 (shifted)')
plt.ylabel('Dimension 1 (shifted)')
plt.show()
```

```python
import nmrglue as ng
import numpy as np # Import numpy to use np.complex64
import os

data_to_save = data_2d_fid

# Create a basic Bruker parameter dictionary
bruker_dic = {} # Start with an empty dictionary

# Attempt to guess parameters from the data
bruker_dic = ng.bruker.guess_udic(bruker_dic, data_to_save)

# Add or update parameters for complex data and dimensions
# These are simplified examples; actual Bruker parameters can be more complex
bruker_dic['procs'] = bruker_dic.get('procs', {}) # Ensure 'procs' key exists
bruker_dic['procs']['DTYPA'] = '_CM'  # Data Type: Complex

# Add dimension sizes, ensuring they are integers
bruker_dic['procs']['SI'] = int(data_to_save.shape[1]) # Size in the direct dimension (usually F2)
bruker_dic['proc2'] = bruker_dic.get('proc2', {}) # Ensure 'proc2' key exists for indirect dimension
bruker_dic['proc2']['SI'] = int(data_to_save.shape[0]) # Size in the indirect dimension (usually F1)

# If the original data was complex and we want to preserve that info in the dictionary
# We might need parameters from the acquisition side as well (acqu/acqus)
bruker_dic['acqu'] = bruker_dic.get('acqu', {})
bruker_dic['acqu']['TD'] = int(data_to_save.shape[1] * 2) # Total Data points (real + imag) in direct dim
bruker_dic['acqus'] = bruker_dic.get('acqus', {})
bruker_dic['acqus']['TD'] = int(data_to_save.shape[0]) # Total Data points in indirect dim
bruker_dic['_comments'] = 'KK' # Total Data points in indirect dim

output_dir = '../processed_nmr_data'
os.makedirs(output_dir, exist_ok=True)

#bruker_dic.pop('_comments')
ng.bruker.write(output_dir, bruker_dic, data_2d_fid, write_acqus=True, write_prog=False, overwrite=True,pdata_folder=True)

#print(bruker_dic)
# Other parameters might be needed depending on the specific Bruker software/version
# For example, BYTORDP (Byte Order), NC (Number of complex points per FID)

#print("Bruker dictionary updated with complex data parameters.")
# Optionally display the dictionary to inspect
# display(bruker_dic)
```
